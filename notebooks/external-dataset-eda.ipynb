```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load the Excel file containing GHG data
ghg_file = "finalquarterlyghgemissions.xlsx"
df_ghg = pd.read_excel(ghg_file, sheet_name="Table 1", skiprows=3)  # Skip first 3 rows for clean headers

# Parse the 'Time Period' column into 'Year' and 'Quarter'
df_ghg["Year"] = df_ghg["Time Period"].apply(lambda x: int(x.split()[0]))
df_ghg["Quarter"] = df_ghg["Time Period"].apply(lambda x: int(x.split()[1][1]))
```

```python
# Visualizing the GHG emissions trend over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Seasonally adjusted emissions', data=df_ghg)
plt.title('GHG Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('GHG Emissions')
plt.grid(True)
plt.show()

# Distribution plot for emissions
plt.figure(figsize=(8, 5))
sns.histplot(df_ghg['Seasonally adjusted emissions'].dropna(), kde=True)
plt.title('Distribution of GHG Emissions')
plt.xlabel('GHG Emissions')
plt.grid(True)
plt.show()
```

```python
# Checking for missing values
missing_values = df_ghg.isnull().sum()
print("Missing Values:\n", missing_values)

# Summary statistics
print(df_ghg.describe())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Seasonally adjusted emissions', hue='Quarter', data=df_ghg, palette='tab10')
plt.title('GHG Emissions Trends by Quarter')
plt.xlabel('Year')
plt.ylabel('GHG Emissions')
plt.grid(True)
plt.show()
```
